### Read Dataset

```python
import pandas as pd
data = pd.read_csv('zomato.csv', encoding='latin1')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Explore categorical variables
categorical_cols = ['Country Code', 'City', 'Cuisines', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Price range', 'Rating text']
for col in categorical_cols:
    print(data[col].value_counts())

# Visualize categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='Rating text', data=data)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

```

```python
# Explore numerical variables
numerical_cols = ['Average Cost for two', 'Aggregate rating', 'Votes']
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import OneHotEncoder

# Selecting features and target variable
X = data[['Country Code', 'City', 'Cuisines', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 'Price range']]
y = data['Rating text']

# Perform one-hot encoding for categorical variables
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(X)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize and train the Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Predictions on the testing set
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import joblib  # Import joblib directly

# Save the trained model
joblib.dump(clf, 'model.pkl')
```
